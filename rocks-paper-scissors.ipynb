```python
import cv2
```

```python
cap = cv2.VideoCapture( 0 )
```

```python
from cvzone.HandTrackingModule import HandDetector
```

```python
detector = HandDetector()
```

```python
import random
import time
you = 0
comp = 0
print("ROCKS PAPERS SCISSORS\n\n\n\n")
while True:
    time.sleep(1)
    status, photo = cap.read()
    cv2.imshow("my photo", photo)
    if cv2.waitKey(100) == 13:
        break
    handphoto = detector.findHands(photo , draw=False)
    print()
    if handphoto:
        lmlist = handphoto[0]
        fingerstatus = detector.fingersUp(lmlist)
        choice = random.randint(1,3)
        
        if fingerstatus == [1,1,1,1,1]:
            print("You chose Papers")
            if(choice == 1):
                print("Computer chose rocks,\n\n YOU WIN!!!")
                you+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))

            elif(choice == 2):
                print("Computer chose papers,\n\n it's a DRAW")
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
            else:
                print("Computer chose scissors,\n\n YOU LOSE!!!")
                comp+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
    
        elif fingerstatus == [ 0 ,1 ,1 , 0, 0]:
            print("You chose scissors")
            if(choice == 1):
                print("Computer chose rocks,\n\n YOU LOSE!!!")
                comp+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
            elif(choice == 2):
                print("Computer chose papers,\n\n YOU WIN!!!")
                you+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
            else:
                print("Computer chose scissors,\n\n it's a DRAW!")
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
    
        elif fingerstatus == [ 0 , 0, 0, 0 , 0 ]:
            print("You chose rock")
            if(choice == 1):
                print("Computer chose rocks,\n\n it's a DRAW")
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
            elif(choice == 2):
                print("Computer chose papers,\n\n YOU LOSE")
                comp+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
            else:
                print("Computer chose scissors,\n\n YOU WIN!!!")
                you+=1
                print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
    
        else:
            print("Invalid Input")
            print("\n\nYour points: " + str(you) + " Computer points: " + str(comp))
cv2.destroyAllWindows()
```
